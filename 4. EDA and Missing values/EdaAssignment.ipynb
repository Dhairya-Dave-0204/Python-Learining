## 4. Relationship Analysis

### Correlation Analysis

A correlation matrix was generated using:

```python
df.corr()
```

and visualized using a heatmap.

### Observations

* Correlation values across the dataset were generally very weak.
* No strong positive correlations were observed.
* No strong negative correlations were observed.
* Most relationships were close to zero, indicating limited linear dependence among variables.

### Scatter Plot Analysis

Scatter plots were created to examine relationships between:

* Price vs KM Driven
* Price vs Car Age
* Price vs Engine CC
* Price vs Horsepower

### Observations

#### Price vs KM Driven

* Points appeared widely dispersed.
* No clear increasing or decreasing trend was visible.
* Some clusters were observed around moderate mileage and moderate price ranges.

#### Price vs Car Age

* Distinct vertical bands appeared due to discrete age values.
* Price variation remained high even within the same vehicle age group.

#### Price vs Engine CC

* The scatter pattern appeared broadly distributed.
* No strong visual trend was evident.

#### Price vs Horsepower

* Several observations showed lower horsepower values associated with higher prices.
* No clear linear relationship was identified.

### Hypothesis Testing

Initial assumptions included:

* Car Age may influence Price.
* KM Driven may influence Price.
* Horsepower and Torque may be related.
* Length and Width may be related.

Analysis showed that several expected relationships were weaker than anticipated, highlighting the importance of validating assumptions using data rather than relying solely on intuition.
