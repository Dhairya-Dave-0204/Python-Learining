# **1. Pandas**

**What is Pandas?**

- Pandas is a Python library used to handle and analyze data.
- It is built on top of NumPy and is especially designed for working with tabular data (rows and columns).
- Think of Pandas as a tool that makes Python work like Excel or Google Sheets, but with more power and automation.

**Why Use Pandas?**

- **Organized Data:** Data is stored in a table format (rows × columns) like spreadsheets.
- **Easy Analysis:** Functions like sum(), mean(), describe() make analysis simple.
- **Real-World Data:** Most datasets in AI/ML are in CSV or Excel — Pandas makes it easy to load them.
- **Better than NumPy:** NumPy arrays only have numbers, but Pandas can handle numbers, text, dates, etc., together.

Pandas provides two main objects:

**2. DataFrame (2D data)**

# **2. Pandas Series (1D Data)**

A Series is like a single column of data in Excel.
It has two parts:

- Index (row labels, default is 0,1,2,…)
- Values (the actual data)

**1. Creating a Series from a List**

```python
import pandas as pd

marks = [90, 85, 78, 92]
s = pd.Series(marks)

print(s)
```

**2. Custom Index**

We can give our own index labels.

```python
marks = [90, 85, 78, 92]
students = ["Amit", "Sara", "John", "Rita"]

s = pd.Series(marks, index=students)

print(s)
```

**3. Creating Series from a Dictionary**

```python
data = {"Amit": 90, "Sara": 85, "John": 78}
s = pd.Series(data)
print(s)
```

**4. Accessing Elements in a Series**

**By index number**

```python
print(s[0])
```

**By label**

```python
print(s["Sara"])
```

**Multiple values**

```python
print(s[["Amit", "John"]])
```

**5. Basic Functions with Series**

```python
print("Sum:", s.sum())
print("Mean:", s.mean())
print("Maximum:", s.max())
print("Minimum:", s.min())
print("First 2 values:\n", s.head(2))
print("Last 2 values:\n", s.tail(2))
```

# **3. Pandas DataFrame (2D Data)**

A DataFrame is like an Excel sheet or a database table.

- It has rows and columns.
- Each column can have a different type of data (numbers, text, dates, etc.).
- Rows have an index (default = 0,1,2,…).

**Creating DataFrame from a Dictionary**

```python
import pandas as pd

data = {
    "Name": ["Amit", "Sara", "John"],
    "Age": [21, 20, 22],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)
print(df)
```

**Creating DataFrame from List of Dictionaries**

```python
students = [
    {"Name": "Amit", "Age": 21, "Marks": 85},
    {"Name": "Sara", "Age": 20, "Marks": 90},
    {"Name": "John", "Age": 22, "Marks": 88}
]

df = pd.DataFrame(students)
print(df)
```

**Creating DataFrame from 2D List**

```python
data = [
    ["Amit", 21, 85],
    ["Sara", 20, 90],
    ["John", 22, 88]
]

df = pd.DataFrame(data, columns=["Name", "Age", "Marks"])
print(df)
```

**Viewing Data**

```python
print("First 2 rows:\n", df.head(2))
print("Last 2 rows:\n", df.tail(2))
print("Shape (rows, cols):", df.shape)
print("Column names:", df.columns)
```

**Summary Information**

```python
print("Info:\n")
print(df.info())


print("\nDescribe:\n")
print(df.describe())
```

**Accessing Columns**

```python
print("Name column:\n", df["Name"])
print("Age column:\n", df["Age"])
```

**Accessing Multiple Columns**

```python
print(df[["Name", "Marks"]])
```

**Accessing Rows**

By index position (iloc)

```python
print(df.iloc[0])
print(df.iloc[1:3])
```

**By label (loc)**

```python
print(df.loc[0])
print(df.loc[[0,2]])
```

# **4. Indexing & Slicing in Pandas DataFrame**

Like NumPy arrays, we can access elements in Pandas using indexes.
But Pandas also allows us to use column names and row labels, making it much easier.

**Selecting a **Single** Column**

```python
import pandas as pd

data = {
    "Name": ["Amit", "Sara", "John"],
    "Age": [21, 20, 22],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)

print("Name column:\n", df["Name"])
```

**Selecting Multiple Columns**

```python
print(df[["Name", "Marks"]])
```

**Selecting Rows by Position (iloc)**

iloc works like normal Python indexing (by position).

```python
print("First row:\n", df.iloc[0])
print("\nFirst two rows:\n", df.iloc[0:2])
```

**Selecting Rows by Label (loc)**

loc uses row labels (index).

```python
print("Row with index 1:\n", df.loc[1])
print("\nRows 0 and 2:\n", df.loc[[0, 2]])
```

**Selecting Specific Rows & Columns**

```python
print(df.loc[0, "Name"])
print(df.loc[0:1, ["Name", "Marks"]])
```

**Conditional Selection (Filtering)**

```python
print("Students with Marks > 85:\n", df[df["Marks"] > 85])
```
