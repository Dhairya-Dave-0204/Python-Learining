# **1. Matplotlib**

**What is Matplotlib?**
- Matplotlib is a Python library used for data visualization.
- It helps us create graphs and charts to understand data better.
- Instead of looking only at numbers in tables, graphs make it easier to see patterns and comparisons.

**Why Use Matplotlib?**

- To make data easy to understand.
- To find trends, patterns, and relationships.
- It is the most popular graphing library in Python (used in Data Science, AI, and Machine Learning).

The most common way is:

```python
import matplotlib.pyplot as plt
```

Here, pyplot is a module inside Matplotlib, and plt is just a short name (alias).

**First Example – A Simple Line Plot**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]   # x-axis values
y = [2, 4, 6, 8, 10]  # y-axis values

plt.plot(x, y)        # create line plot
plt.show()            # display the graph
```

**Adding Labels and Title**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel("X-axis")           # label for x-axis
plt.ylabel("Y-axis")           # label for y-axis
plt.title("Simple Line Graph") # title
plt.show()
```

**Display both graphs one after another.**

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

x1 = [1, 3, 4, 5, 6]
y1 = [4, 9, 16, 25, 36]

# First Graph
plt.plot(x, y)
plt.xlabel("Numbers")
plt.ylabel("Squares")
plt.title("Square Numbers Graph")
plt.show()

# Second Graph
plt.plot(x1, y1)
plt.xlabel("Numbers")
plt.ylabel("Cubes")
plt.title("Cube Numbers Graph")
plt.show()
```

# **2. Basic Plot – Line Styles, Colors, Markers**

In Matplotlib, we can customize the appearance of lines in our plots:

**1. Change Line Color**

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, color="green")   # line color
plt.title("Line with Green Color")
plt.show()
```

**2. Change Line Style**

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y, linestyle="--")   # dashed line
plt.title("Dashed Line")
plt.show()

plt.plot(x, y, linestyle=":")    # dotted line
plt.title("Dotted Line")
plt.show()
```

**3. Add Markers**

Markers show each data point.

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]


plt.plot(x, y, marker="o")    # circle markers
plt.title("Line with Circle Markers")
plt.show()

plt.plot(x, y, marker="s")    # square markers
plt.title("Line with Square Markers")
plt.show()

plt.plot(x, y, marker="*")    # star markers
plt.title("Line with Star Markers")
plt.show()
```

**4. Combine Color, Style, and Marker**

```python
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x,y, color='Green', linestyle='--', marker='o')
plt.title("Green Dashed line with Circle Markers")
plt.show()
```

# **3. Multiple Lines in One Graph**

**1. Plotting Two Lines Together**

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]   # first line
y2 = [1, 4, 9, 16, 25]  # second line

plt.plot(x, y1, label="Line 1")   # first line with label
plt.plot(x, y2, label="Line 2")  # second line with label

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two Lines in One Graph")
plt.legend()   # show legend
plt.show()
```

**2. Using Different Styles for Each Line**

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]   # first line
y2 = [1, 4, 9, 16, 25]  # second line

plt.plot(x, y1, color="blue", linestyle="--", marker="o", label="y=2x")
plt.plot(x, y2, color="red", linestyle=":", marker="s", label="y=x^2")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two Styled Lines")
plt.legend()
plt.show()
```

**3. Plotting More Than Two Lines**

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]   # first line
y2 = [1, 4, 9, 16, 25]  # second line
y3 = [1, 8, 27, 64, 125]  # cube

plt.plot(x, y1, label="Line 1")
plt.plot(x, y2, label="Line 2")
plt.plot(x, y3, label="Line 3")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Multiple Lines in One Graph")
plt.legend()
plt.show()
```

# **4. Bar Charts in Matplotlib**

**What is a Bar Chart?**

- A bar chart is used to represent categorical data with rectangular bars.
- Each bar’s height (or length) shows the value for that category.

**Why Do We Use Bar Charts?**

- To compare categories (like sales of products, marks of students, etc.).
- Helps to visualize differences clearly instead of just looking at numbers.
- Very useful when dealing with discrete data (categories, groups).

**Role of Bar Charts in Data Visualization**

- Shows comparisons between categories.
- Easy to understand for non-technical people (management, clients, students).
- Used widely in business reports, surveys, and presentations.

**1. Vertical Bar Chart**

```python
students = ["Amit", "Sara", "John", "Rita"]
marks = [85, 90, 75, 95]

plt.bar(students, marks, color="skyblue")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Marks of Students")
plt.show()
```

**2. Horizontal Bar Chart**

```python
students = ["Amit", "Sara", "John", "Rita"]
marks = [85, 90, 75, 95]

plt.barh(students, marks, color="green")
plt.xlabel("Marks")
plt.ylabel("Students")
plt.title("Marks of Students (Horizontal)")
plt.show()
```

**3. Customize Bar Colors and Width**

```python
students = ["Amit", "Sara", "John", "Rita"]
marks = [85, 90, 75, 95]

plt.bar(students, marks, color=["red", "blue", "green", "orange"], width=0.5)
plt.title("Customized Bar Chart")
plt.show()
```

**4. Multiple Bar Charts (Side by Side)**

```python
import numpy as np

students = ["Amit", "Sara", "John", "Rita"]
marks1 = [85, 90, 75, 95]  # test 1
marks2 = [80, 88, 70, 92]  # test 2

x = np.arange(len(students))  # positions

plt.bar(x - 0.2, marks1, width=0.4, label="Test 1")
plt.bar(x + 0.2, marks2, width=0.4, label="Test 2")

plt.xticks(x, students)
plt.xlabel("Students")
plt.ylabel("Marks")
plt.title("Comparison of Marks in Two Tests")
plt.legend()
plt.show()
```

# **5. Histogram in Matplotlib**

**What is a Histogram?**

- A histogram is used to show the distribution of data (how data is spread).
- It divides the data into bins (intervals) and shows how many values fall into each bin.
- Looks like a bar chart, but it’s for continuous data (not categories).

**Why Do We Use Histograms?**

- To see patterns in data (normal distribution, skewness, etc.).
- To find frequency of values within ranges.
- Common in statistics, data analysis, machine learning (e.g., distribution of marks, salaries, ages).

**Role of Histograms in Data Visualization**

- Help us understand how values are distributed.
- Useful to detect outliers (extreme values).
- Show if the data is centered, spread out, or skewed.

**1. Basic Histogram**

```python
marks = [45, 55, 67, 89, 76, 45, 90, 82, 56, 71, 62, 50, 65, 77, 85]

plt.hist(marks, bins=5, color="skyblue", edgecolor="black")
plt.xlabel("Marks Range")
plt.ylabel("Number of Students")
plt.title("Distribution of Marks")
plt.show()
```

**2. Changing Number of Bins**



```python
marks = [45, 55, 67, 89, 76, 45, 90, 82, 56, 71, 62, 50, 65, 77, 85]

plt.hist(marks, bins=10, color="orange", edgecolor="black")
plt.title("Histogram with 10 Bins")
plt.show()
```

**3. Histogram with Grid**

```python
marks = [45, 55, 67, 89, 76, 45, 90, 82, 56, 71, 62, 50, 65, 77, 85]

plt.hist(marks, bins=5, color="green", edgecolor="black")
plt.grid(True)
plt.title("Histogram with Grid")
plt.show()
```

# **6. Scatter Plot in Matplotlib**

**What is a Scatter Plot?**

- A scatter plot is used to show the relationship between two variables.
- Each point on the graph represents a pair of values (x, y).

**Why Do We Use Scatter Plots?**

- To check if there is a correlation (connection) between two variables.
- To see patterns, clusters, or outliers.
- Example: height vs weight, study hours vs marks.

**Role of Scatter Plots in Data Visualization**

- Show whether two variables are positively related, negatively related, or have no relation.
- Useful in data science, machine learning, and statistics.
- Makes it easy to spot groups or unusual points.

**1. Basic Scatter Plot**

```python
x = [5, 7, 8, 7, 6, 9, 5, 6, 7, 8]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color="blue")
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Scatter Plot Example")
plt.show()
```

**2. Customize Scatter Plot**

```python
x = [5, 7, 8, 7, 6, 9, 5, 6, 7, 8]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]

plt.scatter(x, y, color="red", marker="*", s=50) # s is size of point
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.title("Customized Scatter Plot")
plt.show()
```

**3. Two Groups in One Scatter Plot**

```python
# Group 1
x1 = [5, 6, 7, 8]
y1 = [80, 82, 85, 90]

# Group 2
x2 = [6, 7, 8, 9]
y2 = [60, 62, 65, 70]

plt.scatter(x1, y1, color="blue", label="Group 1")
plt.scatter(x2, y2, color="red", label="Group 2")

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Two Groups in Scatter Plot")
plt.legend()
plt.show()
```

# **7. Pie Chart in Matplotlib**

**What is a Pie Chart?**

- A pie chart shows data as slices of a circle (like pieces of a pizza).
- Each slice represents a proportion (percentage) of the total.

**Why Do We Use Pie Charts?**

- To show part-to-whole relationships.
- Easy to understand for comparing percentages.
- Example: Expenses distribution, market share, student grades.

**Role of Pie Charts in Data Visualization**

- Useful when showing percentages or proportions.
- Gives a quick overview of how a total is divided.
- Common in business reports, surveys, and presentations.

**1. Basic Pie Chart**

```python
expenses = [3000, 1500, 1200, 800]
categories = ["Rent", "Food", "Travel", "Others"]

plt.pie(expenses, labels=categories)
plt.title("Monthly Expenses")
plt.show()
```

**2. Pie Chart with Percentages**

```python
expenses = [3000, 1500, 1200, 800]
categories = ["Rent", "Food", "Travel", "Others"]

plt.pie(expenses, labels=categories, autopct="%1.1f%%")
plt.title("Monthly Expenses with Percentage")
plt.show()
```

**3. Exploding a Slice (Highlighting)**

```python
expenses = [3000, 1500, 1200, 800]
categories = ["Rent", "Food", "Travel", "Others"]

plt.pie(expenses, labels=categories, autopct="%1.1f%%", explode=[0.1, 0, 0, 0])
plt.title("Monthly Expenses (Rent Highlighted)")
plt.show()
```

**4. Adding Colors**

```python
expenses = [3000, 1500, 1200, 800]
categories = ["Rent", "Food", "Travel", "Others"]

colors = ["red", "blue", "green", "orange"]
plt.pie(expenses, labels=categories, autopct="%1.1f%%", colors=colors)
plt.title("Monthly Expenses with Colors")
plt.show()
```
