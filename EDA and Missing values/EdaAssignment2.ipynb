#### Based on the above analysis the filling of missing values will be as follows:

Numerical Columns:
- age                -> Mean
- salary             -> Mean
- training_hours     -> Mean
- performance_score  -> Mean
- experience_years   -> Mean
- projects_completed -> Mean

Categorical Columns:
- department              -> Unknown
- education               -> Unknown
- job_role                -> Unknown
- work_mode               -> Unknown
- promotion_last_5yrs     -> Unknown
- attrition               -> Unknown

This is done because the numerical values have no possible outliers and data is not skewed so mean is better choice , inverse is true for median

For the categorical columns if mode is used than data will become one category dominant so a new category Unknown is created

# Employee Attrition Dataset - Exploratory Data Analysis (EDA) Report

## 1. Introduction

### Objective

The objective of this Exploratory Data Analysis (EDA) is to understand the employee attrition dataset, identify data quality issues, analyze variable distributions, investigate relationships between variables, and discover factors that may influence employee attrition.

---

## 2. Dataset Overview

### Dataset Description

The dataset contains employee-related information including:

* Age
* Experience Years
* Salary
* Training Hours
* Performance Score
* Projects Completed
* Department
* Education
* Job Role
* Work Mode
* Promotion Status
* Attrition Status

### Initial Observations

* Dataset contains both numerical and categorical features.
* Numerical columns were stored as floating-point values.
* Certain variables such as Age and Projects Completed contained continuous decimal values.
* Missing values were present in multiple numerical and categorical columns.

---

## 3. Data Quality Assessment

### Missing Values

Missing values were identified in:

#### Numerical Columns

* Age
* Experience Years
* Salary
* Training Hours
* Performance Score
* Projects Completed

#### Categorical Columns

* Department
* Education
* Job Role
* Work Mode
* Promotion Last 5 Years
* Attrition

### Duplicate Records

* No duplicate records were found in the dataset.

---

## 4. Missing Value Treatment

### Numerical Features

Histograms and boxplots were analyzed before selecting an imputation strategy.

#### Observations

* Distributions were approximately balanced.
* No significant skewness was observed.
* No major outliers were detected.
* Median values were generally centered.

#### Action Taken

Mean imputation was applied to all numerical columns.

### Categorical Features

Categorical variables contained a substantial number of missing values.

Using mode imputation would have artificially increased the frequency of a single category and distorted the distribution.

#### Action Taken

Missing categorical values were replaced with a new category:

```text
Unknown
```

This preserved information regarding missingness while maintaining the original category distributions.

## 10. Key Insights

### Data Quality Insights

* Missing values were present across multiple columns.
* Numerical distributions were suitable for mean imputation.
* Large categorical missingness was preserved using an Unknown category.

### Attrition Insights

* Age does not appear strongly associated with attrition.
* Salary does not appear strongly associated with attrition.
* Work mode does not appear strongly associated with attrition.
* Promotion history does not appear strongly associated with attrition.
* Department showed the strongest relationship with attrition.
* Sales employees were more likely to stay than leave.
* Job role showed only minor variation in attrition behavior.

### Relationship Insights

* Correlation analysis revealed negligible relationships among numerical variables.
* Multivariate analysis showed no meaningful clustering or separation among attrition groups.

---

## 11. Final Conclusion

The dataset exhibits generally weak relationships among variables. Most numerical and categorical features showed little ability to distinguish employees who left from those who stayed. Department was the only variable showing a noticeable difference, with the Sales department exhibiting lower attrition than other departments.

Overall, the dataset appears highly balanced and contains limited evidence of strong predictors of employee attrition. The analysis suggests that attrition behavior in this dataset is not strongly explained by the available variables.
