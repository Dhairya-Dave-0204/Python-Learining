7. Write a program to swap two numbers using a temporary variable. 

```python
a = 10
b = 20
print("Original values\n A = " + str(a) + " B = " + str(b))
# Swap logic
temp = a
a = b
b = temp

print("Swapped values\n A = " + str(a) + " B = " + str(b))
```
