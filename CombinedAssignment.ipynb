Q40. Generate synthetic temperature data using NumPy, store it in Pandas, and visualize it using Matplotlib.

```python
tempData = np.random.randint(15, 40, 7)
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]

tempratureData = {
    "day": days,
    "temprature": tempData
}

df = pd.DataFrame(tempratureData)

print(df)

plt.plot(tempratureData["day"], tempratureData["temprature"])
plt.xlabel("Days")
plt.ylabel("Temprature")
plt.show()
```
