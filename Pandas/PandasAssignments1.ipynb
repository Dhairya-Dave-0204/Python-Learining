1. Create a Series from the list [3, 6, 9, 12] with index ['a','b','c','d'].

```python
myList = [3, 6, 9, 12]
counter = ['a','b','c','d']

s = pd.Series(myList, index=counter)

print(s)
```

4) Replace negative values in the Series s = pd.Series([4, -3, 7, -1, 2]) with 0.

```python
u = pd.Series([4, -3, 7, -1, 2])

print(u.where(u > 0, 0))
```

5) Create a Series from the dictionary {'pen': 10, 'pencil': 5, 'eraser': 3} and select values >= 5.

```python
original = {'pen': 10, 'pencil': 5, 'eraser': 3}

ds = pd.Series(original)

print(ds[ds >= 5])
```

6) Given s = pd.Series([2, 4, 6, 8]), find mean, max, and describe().

```python
v = pd.Series([2, 4, 6, 8])

print("Mean: ", v.mean())
print("Max: ", v.max())
print("Describe: \n",v.describe())
```
