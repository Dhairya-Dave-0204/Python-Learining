# Customer Churn Dataset - Complete EDA

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('customer_churn_realistic.csv')
df.head()
```

```
# 1. Overview
df.shape
df.info()
```

```
# 2. Summary Statistics
df.describe(include='all')
```

```
# 3. Missing Values
df.isnull().sum()
```

```
# 4. Duplicates
df.duplicated().sum()
```

```
# 5. Separate Numerical & Categorical Columns
num_cols=df.select_dtypes(include='number').columns
cat_cols=df.select_dtypes(include='object').columns
```

```
# 6. Missing Value Handling
for c in num_cols:
    if df[c].isnull().sum()>0:
        df[c]=df[c].fillna(df[c].median())
for c in cat_cols:
    if df[c].isnull().sum()>0:
        df[c]=df[c].fillna(df[c].mode()[0])
```

```
# 7. Histograms
for c in num_cols:
    df[c].hist(); plt.title(c); plt.show()
```

```
# 8. Boxplots
for c in num_cols:
    sns.boxplot(x=df[c]); plt.title(c); plt.show()
```

```
# 9. Categorical Analysis
for c in cat_cols:
    df[c].value_counts().plot(kind='bar'); plt.title(c); plt.show()
```

```
# 10. Correlation Heatmap
sns.heatmap(df[num_cols].corr(),annot=True,cmap='coolwarm')
plt.show()
```

```
# 11. Churn Analysis
# Replace column names if needed based on dataset
if 'churn' in df.columns:
    sns.countplot(x='churn',data=df)
    plt.show()
```
